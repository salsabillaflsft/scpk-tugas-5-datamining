```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DataRekomendasi.csv') 
df
```

```python
Ya = df[df["Rekomendasi"] == "Ya"]
Tidak = df[df["Rekomendasi"] == "Tidak"]

data = (Ya, Tidak)
colors = ("green","red")
groups = ("Ya", "Tidak")

for data, color, group in zip(data, colors, groups):
    x, y = data["Kehadiran"], data["Prakarsa"]
    plt.scatter(x, y, c=color, label=group)
plt.legend()
plt.title("Plot")
plt.xlabel("Kehadiran")
plt.ylabel("Prakarsa")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = df.iloc[:, :4]
y = df.iloc[:, 4]
print(X)
print(y)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X,y)
```
